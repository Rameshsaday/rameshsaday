```python
# univarient analysis

#frequency table for age

ft = ds['number of bedrooms'].value_counts()

print("Frequency table for No of bedrooms is given below")
print("{}\n\n\n".format(ft))



print("Mean,Mode, Median, std \n")
ma=ds['living area'].mean() #mean of living area
mh = ds['number of floors'].mode() #mode of number of floors
mel = ds['number of bathrooms'].median() #median of number of bathrooms
stw = ds['condition of the house'].std() #standard devation of condition of the house

print("1-mean of living area = ",ma) 
print("2-mode of number of floors = ",mh) 
print("3-median number of bathrooms= ",mel)#
print("4-standard devation of condition of the house = ",stw) 
```

```python
import seaborn as sns
from matplotlib import pyplot as plt


sns.relplot(data=ds,x='number of bedrooms',y='Price')
sns.set_context('notebook')

sns.relplot(data=ds,x='number of bedrooms',y='Price',hue='Number of schools nearby')


sns.relplot(data=ds,x='number of bedrooms',y='Price',hue='condition of the house',col='Number of schools nearby',col_wrap=2)
```

```python
#multi-varient analysis

sns.countplot(x='number of bedrooms',hue='Number of schools nearby',data=ds,palette='magma')
```
